```python
import pandas as pd
```

```python
ml = pd.read_csv('Meteorite_Landings (1).csv')
ml.head()
```

```python
ml.drop('nametype', inplace = True, axis = 1)
ml.head()
```

```python
ml[ml['mass (g)'].isna()]
ml['mass (g)'].fillna(0,inplace=True)
len(ml[ml['mass (g)'].isna()])
ml.dtypes
```

```python
ml['decade'] = ml.year//10*10
decade = ml.decade.unique()
with pd.ExcelWriter('MeteoriteLandingsByDecade.xlsx') as xlsxWriter:
    for d in decade:
        ml[ml.decade == d].to_excel(xlsxWriter,sheet_name=str(d))
xlsxWriter.save()
xlsxWriter.close()
```
